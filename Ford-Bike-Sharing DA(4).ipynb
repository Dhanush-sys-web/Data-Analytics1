## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# Impute missing values with median (example for a column 'duration_sec')
df['duration_sec'].fillna(df['duration_sec'].median(), inplace=True)

# 2. Converting Data Types
# Convert 'start_time' to datetime and 'user_type' to categorical
df['start_time'] = pd.to_datetime(df['start_time'])
df['user_type'] = df['user_type'].astype('category')

# 3. Handling Outliers
# Remove trips with duration greater than 10,000 seconds
df = df[df['duration_sec'] < 10000]

# 4. Filtering Data
# Filter data for 'Subscribers' only
df_subscribers = df[df['user_type'] == 'Subscriber']

# **Instead of filtering for an entire year (2018), filter for January 2018 as your data suggests:**
df = df[(df['start_time'] >= '2018-01-01') & (df['start_time'] <= '2018-01-31')]

# If you still get the error, comment out ALL filtering and identify which filter is causing the issue:

# df = df[df['duration_sec'] < 10000]  # comment to test
# df_subscribers = df[df['user_type'] == 'Subscriber']  # comment to test
# df = df[(df['start_time'] >= '2018-01-01') & (df['start_time'] <= '2018-01-31')]  # comment to test


# 5. Creating New Columns
# Extract hour from 'start_time' and day of the week
df['hour'] = df['start_time'].dt.hour
df['day_of_week'] = df['start_time'].dt.dayofweek

# Calculate age from birth year
current_year = 2025
# Check if 'member_birth_year' column exists in your DataFrame, if not use the name of the correct column
if 'member_birth_year' in df.columns:
    df['age'] = current_year - df['member_birth_year']
elif 'birth_year' in df.columns:
    df['age'] = current_year - df['birth_year']
else:
    print("Column for birth year not found.")

# 6. Removing Duplicates
df.drop_duplicates(inplace=True)

# 7. Renaming Columns
df.rename(columns={'start_station_name': 'Start Station', 'end_station_name': 'End Station'}, inplace=True)

# Before using MinMaxScaler, install and import it
!pip install scikit-learn
from sklearn.preprocessing import MinMaxScaler

# 8. Normalizing or Scaling Data (Example: Normalizing 'duration_sec')
# Check if df is empty before scaling
if df.empty:
    print("DataFrame is empty after filtering. Check your filters.")
else:
    scaler = MinMaxScaler()
    df['normalized_duration'] = scaler.fit_transform(df[['duration_sec']])

# 9. Categorizing Data (Example: Categorizing 'age' into age groups)
bins = [0, 18, 35, 60, 100]
labels = ['<18', '18-35', '35-60', '60+']
df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels)

# Display the first few rows of the cleaned dataset
df.head()
````
